## foliumにmethodを追加
既存のClassに、こんな方法でmethodを追加できるなんて、知らなかった。

```
# Define a method for displaying Earth Engine image tiles to folium map.
def add_ee_layer(self, ee_image_object, vis_params, name):
  map_id_dict = ee.Image(ee_image_object).getMapId(vis_params)
  folium.raster_layers.TileLayer(
    tiles = map_id_dict['tile_fetcher'].url_format,
    attr = 'Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    name = name,
    overlay = True,
    control = True
  ).add_to(self)

# Add EE drawing method to folium.
folium.Map.add_ee_layer = add_ee_layer
```

##Image出力関数定義その２
image objectから取得できるcoordinatesの外枠をregionに定義</br>
画像は、google driveの"skysat"に出力</br>
scaleについてはこちら。</br>
最精細1~500荒い</br>
https://developers.google.com/earth-engine/scale

```
def ImageExportByImageObj(img_obj):
    img_info    =   img_obj.getInfo()
    description =   img_obj.get('system:index').getInfo()
    folder      =   "skysat"
    coords = np.array(img_info["properties"]["system:footprint"]["coordinates"])
    xmax , ymax = np.max(coords,axis=0)
    xmin , ymin = np.min(coords,axis=0)
    #xmin,ymin,xmax,ymax = [-122.44,38.66,-122.45,38.67]
    region  =   ee.Geometry.Rectangle([xmin,ymin,xmax,ymax])
    task = ee.batch.Export.image.toDrive(img_obj,description=description,folder=folder,region=region,scale=1)
    task.start()
```

```
ImageExportByImageObj(img)
```

## 衛星画像と地図の重ね合わせ

```
# Get image coordinates
coords = np.array(img_info["properties"]["system:footprint"]["coordinates"])
coords
# Center of gravity
clon, clat = np.average(coords, axis=0)

# Create a folium map object. (height: size of the window)
my_map = folium.Map(location=[clat, clon], zoom_start=15, height=500)

# Add the elevation model to the map object.
# my_map.add_ee_layer(dem.updateMask(dem.gt(0)), vis_params, 'DEM')
my_map.add_ee_layer(rgb, rgbvis, "RGB")

# Add a layer control panel to the map.
my_map.add_child(folium.LayerControl())

# Display the map.
display(my_map)
```
